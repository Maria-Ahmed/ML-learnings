
# Correlation with Heatmap
- A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types. 
- Very important in Feature Selection
- Heatmap gives visualization
- Helps to see which are positively correlated and which are negatively

```python
df.dtypes
```

```python
df.corr()
```

```python
#just need to provide metrics to the function
sns.heatmap(df.corr())
```

# JoinPlot
A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot



```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
#notice the concentration of points and 
```

```python
# with reg, draws a probability, density functiondistribution curve on histogram, will also draw best fit (regression line)

sns.joinplot(x='tip', y= 'total_bill', data= df, kind= 'reg')
```

# Pair plot
- Also known as a scatterplot
- One variable in the same data row is matched with another variable's value. Shows all variables paired with all the other variables: must be float or int
- Permutation and Combination Type

```python

sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
# Gives color to the field, w.r.t other features and itself
```

```python
#need to see count of gender
df['sex'].value_counts()
```
