```python
# Sine curve
np.pi
```

```python
x= np.arange(0, 4* np.pi, 0.1) # taking array, from 0 to that much value, third parameter: the step
y=np.sin(x)  #taking all values from x values
plt.title('Sine Wave Form')

plt.plot(x,y)
plt.show()
```

```python
# making both cos and sine together
# compute x, y coordinates
x= np.arange(0, 5* np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)


#setup a subplot grid that has height of 2 and width 1, set first subplot as active (1 position pe)
plt.subplot(2,1,1)

#make first plot
plt.plot(x, y_sin)
plt.title('Sine')

plt.subplot(2,1,2)
plt.plot(x, y_cos, 'r--')
plt.title('Cosine')

plt.show()
```

# Box Plot using Matplotlib
 - Helps with the percentile: 0-25-50-75-100
 - Select the normal distribution
 - First two parameters are range, then steps
 - Circles are basically outliers , random selection se at times they also get selected
 - Seaborn better for this
 - Vertical side, and patch se color ayega

```python

data = [np.random.normal(0, std, 100) for std in range(1, 4)]
# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);
```

# Pie Chart


```python
# Data to plot
labels = 'Burger', 'Pizza', 'Roll', 'Chinese'
sizes= [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```
